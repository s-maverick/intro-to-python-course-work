It will produce for you a Python list, called `emails` that contain 1,200,000 emails from three email providers: `amail`, `bmail`, and `cmail`.

````python
import requests
import zipfile
import io

def extract_emails_from_zip_url(url):
    response = requests.get(url)
    zip_file = zipfile.ZipFile(io.BytesIO(response.content))
    txt_file_name = next(name for name in zip_file.namelist() if name.endswith('.txt'))
    with zip_file.open(txt_file_name) as file:
        content = file.read().decode('utf-8')
    emails = [line.strip() for line in content.split('\n') if line.strip()]
    return emails

url = "https://github.com/omerfyalcin/colab-data/raw/main/emails.zip"
emails = extract_emails_from_zip_url(url)
````

## Question 4 (3 points)

Write a function called **get_containing_intervals** that takes a list (call this list **L**) of tuples of length two---each representing an interval---and an integer (call this **integer**). For example, the tuple (5,10) stands for the interval 5 to 10, **including** 5, 10, and all values between them. The function should return the number of intervals that contain the given integer. The tuples may not always be in the "correct" order, the first item in the tuple might be the smaller number (where the interval begins) or the larger one (where it ends).

Test:

```
L = [(10, 2), (-100,0), (-32, 10), (0,7)]
integer = 7
get_containing_intervals(L, integer)
```
should return:

```
3
```

because out of the 4 tuples, only (-100, 0) does not contain 7. Note that `(0, 7)` does contain 7 and is counted towards the 3.

````python
# Write your solution here
# The phrase "pass" is just a placeholder. You can remove that and write your solution in that area

def get_containing_intervals(L, integer):
    count = 0
    for i,j in L:
      mn = min(i,j)
      mx = max(i,j)
      if mn <= integer <= mx:
        count += 1
    print(count)

L = [(10, 2), (-100,0), (-32, 10), (0,7)]
integer = 7
get_containing_intervals(L, integer)
````
